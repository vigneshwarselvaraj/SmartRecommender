```python
import pandas as pd
reviews_df = pd.read_csv('fnl_tempe_with_date.csv',
                      names = ['business_id', 'date', 'star', 'user_id'])
reviews_df.head()
```

```python
reviews_df['year'] = reviews_df['date'].apply(lambda d: int(d[:4]))
last_year_business = reviews_df.groupby('business_id')[['year']].apply(lambda df: df.max()['year'])
```

```python
reviews_df = reviews_df.set_index('business_id')
last_year_business.name = 'last_year'

business_reviews = reviews_df.join(last_year_business)
business_reviews.head()
```

```python
dataset = business_reviews[business_reviews.apply(lambda row: row['year']<row['last_year'], 1)]
dataset.to_csv('yelp_academic_dataset_reviews_for_training.csv')
dataset = business_reviews[business_reviews.apply(lambda row: row['year']==row['last_year'], 1)]
dataset.to_csv('yelp_academic_dataset_reviews_for_labeling.csv')
```
